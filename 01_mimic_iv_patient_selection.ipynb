```python
import pandas as pd
import numpy as np
```

```python
sepsis_patients = pd.read_csv("sepsis3.csv")
```

```python
sepsis_patients['antibiotic_time'] = pd.to_datetime(sepsis_patients['antibiotic_time'], errors='coerce')
sepsis_patients['culture_time'] = pd.to_datetime(sepsis_patients['culture_time'], errors='coerce')
sepsis_patients['suspected_infection_time'] = pd.to_datetime(sepsis_patients['suspected_infection_time'], errors='coerce')
sepsis_patients['sofa_time'] = pd.to_datetime(sepsis_patients['sofa_time'], errors='coerce')

sepsis_patients["sepsis3_onset"] = sepsis_patients[["suspected_infection_time", "sofa_time"]].min(axis=1)
```

```python
file_path = "mimic-iv-2.2/hosp/patients.csv.gz"
patients = pd.read_csv(file_path)
```

```python
patients.head()
```

```python
file_path = "mimic-iv-2.2/icu/icustays.csv.gz"
icu_stays = pd.read_csv(file_path)
```

```python
icu_stays.head()
```

```python
file_path = "mimic-iv-2.2/hosp/admissions.csv.gz"
admissions = pd.read_csv(file_path)
```

```python
admissions.head()
```

```python
admissions["admittime"] = pd.to_datetime(admissions["admittime"])

adult_patients = patients[patients["anchor_age"] >= 18]

merged = pd.merge(adult_patients, admissions, on="subject_id")
merged = pd.merge(merged, icu_stays, on=["subject_id", "hadm_id"])

del patients
del icu_stays
del admissions
```

```python
merged.head()
```

Pacientes com sepse

```python
sepsis_patients = pd.merge(sepsis_patients, merged, on=["subject_id", "stay_id"])
sepsis_patients.to_csv(f"sepsis_patients.csv", index=False)
```

Pacientes sem sepse

```python
no_sepsis_patients = merged[~merged['hadm_id'].isin(sepsis_patients['hadm_id'])]
```

```python
no_sepsis_patients['subject_id'].count()
```

```python
file_path = "mimic-iv-2.2/hosp/diagnoses_icd.csv.gz"
diagnoses = pd.read_csv(file_path)

sepse_codes = {
    9: ["038", "9959", "78552"],
    10: ["A40", "A41", "P36", 'R652'],
    11: ["1G4", '293771399'],
}

sepsis_cases = diagnoses[
    diagnoses.apply(lambda row: any(row["icd_code"].startswith(code) for code in sepse_codes.get(row["icd_version"], [])), axis=1)
]

sepsis_hadm_ids = sepsis_cases["hadm_id"].unique()

no_sepsis_patients = no_sepsis_patients[
    ~no_sepsis_patients["hadm_id"].isin(sepsis_hadm_ids)
]

del diagnoses
del sepsis_cases
del sepsis_hadm_ids
del merged
```

```python
no_sepsis_patients.count()
```

```python
no_sepsis_patients.to_csv(f"no_sepsis_patients.csv", index=False)
```

```python
common_hadm_ids = no_sepsis_patients['stay_id'].isin(sepsis_patients['stay_id'])
common_hadm_rows = no_sepsis_patients[common_hadm_ids]

if not common_hadm_rows.empty:
    print(f"Existem {len(common_hadm_rows)} linhas com 'hadm_id' em comum entre os DataFrames.")
else:
    print("Não há 'stay_id' em comum entre os DataFrames.")

common_subject_ids = no_sepsis_patients['subject_id'].isin(sepsis_patients['subject_id'])
common_subject_rows = no_sepsis_patients[common_subject_ids]

print(f"Existem {len(common_subject_rows)} linhas com 'subject_id' em comum entre os DataFrames.")
```
